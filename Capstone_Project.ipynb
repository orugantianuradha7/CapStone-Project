####**Data types-Statistically**

Data can be either **qualitative** or **quantitative**.

Qualitative data is further divided into **Nominal scale** and **Ordinal scale.**
Quantitative data can be measured on a **continuous scale** or **discrete scale.**

In our data set, there are 4 Nominal variables, 1 Date variable, and 5 continuous variables.

**Qualitative Variables**

1.Name (It is only used as a label, doesn't have any numerical meaning. Statistical operations performed on it doesn't provide significant interpretation. It is measured on a **Nominal Scale**)

2.Country (Names without an order falls under Nominal scale. It is measured on a **Nominal Scale**)

3.Year (It is only used as a label, doesn't have any numerical meaning. Statistical operations performed on it doesn't provide significant interpretation. It is **ordinal**)

4.Sport (Names without an order falls under Nominal scale. It is measured on a **Nominal Scale**)

**Quantitative Variables**

1.Age (It is measurable and can take intermediate values within a range. For example, age of a person can be 12.2 years, which means 12 years and 2 months. It is thus measured on a **continuous scale**)

2.Gold_medal (It is measurable and can't take intermediate values. For example,there can't be 2.8 gold medals. It is thus measured on a **discrete scale**)

3.Silver medal (It is measurable and can't take intermediate values. For example,there can't be 3.5 silver medals. It is thus measured on a **discrete scale**)

4.Bronze medal (It is measurable and can't take intermediate values. For example,there can't be 4.1 bronze medals. It is thus measured on a **discrete scale**)

##**Q2. Dealing with missing values appropriately. You can either remove them or fill them, but a proper justification is required.**

We have seen that 'Name' and 'Age' attributes have 5 null values each.

'Name' is **qualitative** in nature whereas 'Age' is a **quantitative** variable.

We have an option to either replace them with appropriate statistics or drop them altogether.

Before deciding that, it's important to know the **coefficient of skewness** of numerical attribute 'Age'



````python
# Using skew function to find coefficient of skewness and rounding it to 2 decimal places.

round(olympic['age'].skew(),2)
````

Since coefficient of skewness is greater than 0, we say that 'Age' variable is **positively skewed**. The strength of skewness is Strong. Thus, **Median is an appropriate measure of central tendency**. Replacing the missing age values by median age.

###**Transformation of skewed data to Normal/symmetric form**

There are 2 possible methods namely sqrt and log transformation we can try.

Let us check which one gives better transformation and proceed accordingly.

**1. Sqrt method**

````python
# First, we are creating a copy of pandas dataframe 'df1' to add a new column 'Sqrt_age' 
df1 = olympic.copy()

# Using loc method as we are making use of column name instead of index.
df1.insert(len(df1.columns), 'Sqrt_age', np.sqrt(df1.loc[:,"age"]))
````

````python
skew2= round(df1['Sqrt_age'].skew(),2)
print(f"Skewness of Sqrt_age variable is {skew2}")
````

Since the coefficient of skewness is 0.49, the transformation worked.

**2. Log method**

````python
# First, we are creating a copy of pandas dataframe 'df2' to add a new column 'log_age' 
df2 = olympic.copy()

# Using loc method as we are making use of column name instead of index.
df2.insert(len(df2.columns), 'log_age', np.log(df2.loc[:,"age"]))
````

````python
skew3= round(df2['log_age'].skew(),2)
print(f"Skewness of Log age variable is {skew3}")
````

Since the coefficient of skewness is 0.15, it has shown significant drop.

Finally, as **Log Transformation gave great results, we consider it to be the best transformation**

##**Q7. Please perform appropriate numerical measures on each column and report them accordingly.**

To know appropriate statistical measures, we need to know the distribution of variables.

For this, we are writing below code to find the coefficient of skewness for all numerical variables.

````python
# Firstly, we are storing the required column names in a list and assigning it to a variable called "skewness".
skewness = ["age","gold_medal","silver_medal","brone_medal","total_medal"]

# Without running the same code repeatedly, we are making use of a "For Loop" with "if, elif and else" conditional statements.
# round function is used to round off the skewness value to 2 decimal places.

for i in skewness:
   if olympic[i].skew() > 0  :
      print(f"{i} is Positively skewed.\n It has coefficient of skewness=  {round(olympic[i].skew(),2)}\n")
   elif olympic[i].skew() < 0:
      print(f"{i} is Negatively skewed.\n It has coefficient of skewness = {round(olympic[i].skew(),2)}\n")
   else :
      print(f"{i} is Symmetrical.\n It has coefficient of skewness = {round(olympic[i].skew(),2)}\n")
````

**Appropriate Statistical measures:**

For a **Symmetrical data**, Mean is the appropriate measure of central tendency and standard deviation is considered as the appropriate measure of dispersion.

Whereas, for a **Skewed data**, Median is the appropriate measure of central tendency and Inter quartile range(IQR) is considered as the appropriate measure of dispersion.

2.gold_medal

##**Q8. Please list the top 10 countries with the gold medal, then the top10 countries with the silver medal, then the bronze medal and the total medal.**

**a. Top 10 countries (gold medal)**

````python
# We are using 'groupby' function to group the data country wise.
# Additionally, instead of selecting all columns, we only got country and gold_medal column
# Here, we used aggregate function 'sum' on gold_medal attribute to retrieve the total number of gold medals for respective countries.
# Finally, the records are sorted in descending order as per the gold_medal column.

a= olympic[["country","gold_medal"]].groupby("country").sum("gold_medal")
a.sort_values(by=["gold_medal"],inplace=True, ascending= False)
a.rename(columns = {'gold_medal':'Total_gold_medals'}, inplace = True)
top_10_gold= a.head(10)
top_10_gold
````

Thus, we can say that the **United States** received the highest number of gold medals among the rest.

Moreover, 'Russia', 'China', 'Germany', 'Canada', 'Australia', 'Great Britain', 'South Korea' , 'France' and 'Netherlands' are in the top 10 list.

###**Q9. Who was the oldest athlete in the olympics?**

````python
from pyspark.sql.functions import max, desc
````

````python
# Since we need the result for each athlete , groupby function is used.
# In addition to it, we made use of the aggregate function 'max' to get the maximum age

e = spark_df.groupBy('name').max('age').sort(desc('max(age)'))
e.limit(5).show()
````

Hence, **Ian Millar** is the oldest athlete
